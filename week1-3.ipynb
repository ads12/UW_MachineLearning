```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
def words_count_awesome(dict):
    if 'awesome' in dict:
        return dict['awesome']
    else:
        return 0     
    
def words_count_great(dict):
     if 'great' in dict:
        return dict['great']
     else:
        return 0

def words_count_amazing(dict):
     if 'amazing' in dict:
        return dict['amazing']
     else:
        return 0
    
def words_count_fantastic(dict):
    if 'fantastic' in dict:
        return dict['fantastic']
    else:
        return 0

def words_count_love(dict):
     if 'love' in dict:
        return dict['love']
     else:
        return 0
    
def words_count_horrible(dict):
    if 'horrible' in dict:
        return dict['horrible']
    else:
        return 0

def words_count_bad(dict):
     if 'bad' in dict:
        return dict['bad']
     else:
        return 0
    
def words_count_awful(dict):
     if 'awful' in dict:
        return dict['awful']
     else:
        return 0
    
def words_count_wow(dict):
     if 'wow' in dict:
        return dict['wow'] 
     else:
        return 0
    
def words_count_hate(dict):
    if 'hate' in dict:
        return dict['hate']
    else:
        return 0

def words_count_terrible(dict):
    if 'terrible' in dict:
        return dict['terrible']  
    else:
        return 0
```

```python
dict ={'and': 5L, 'stink': 1L,'because': 1L, 'order':7L, 'awesome':30L}
```

```python
products['awesome'] =products['words_count'].apply(words_count_awesome)
products['great'] =products['words_count'].apply(words_count_great)
products['amazing'] =products['words_count'].apply(words_count_amazing)
products['fantastic'] =products['words_count'].apply(words_count_fantastic)
products['love'] =products['words_count'].apply(words_count_love)
products['horrible'] =products['words_count'].apply(words_count_horrible)
products['bad'] =products['words_count'].apply(words_count_bad)
products['awful'] =products['words_count'].apply(words_count_awful)
products['wow'] =products['words_count'].apply(words_count_wow)
products['hate'] =products['words_count'].apply(words_count_hate)
products['terrible'] =products['words_count'].apply(words_count_terrible)
```
